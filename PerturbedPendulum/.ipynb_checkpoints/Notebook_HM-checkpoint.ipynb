```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import random as rd
import time

sys.path.append("..")
tf.random.set_seed(1234)


from AllMethods import ReversibleSymplecticNN as rs
from AllMethods import PerturbedPendulum_Methods as pp

gpu_options = tf.compat.v1.GPUOptions(per_process_gpu_memory_fraction=0.333)
sess = tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(gpu_options=gpu_options))
```

## Load dataset

```python
x, y = pp.read_dataset('x_train.txt', 'y_train.txt')

train_dataset, val_dataset = pp.train_dataset(x[:60], y[:60], 10, 50, 10) #val_len, train_batch, val_batch
```

## Define the model: 
### the first argument is the number of Henon maps, the second argument can be either 'reversible' or 'non-reversible' to generate a model that is either a time-reversible composition of Henon maps or a random composition of Henon maps respectively.

```python
model1R = rs.Henon(15, 'reversible')
```

## Callback function for decaying learning rate

```python
def scheduler(epoch, lr):
    if epoch%100==0:
        print("epoch: ", epoch)
    if epoch < 5:
        return lr
    else:
        return lr*tf.math.exp(-0.0007)
    
callback = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

## Checkpoints

```python
checkpoint_path_1R = "checkpoints_PP_HH_1R_01/cp.ckpt"
checkpoint_dir_1R = os.path.dirname(checkpoint_path_1R)

cp_callback_1R = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path_1R,
                                                 save_weights_only=True,
                                                 save_best_only=True,
                                                 verbose=0)
```

## Training

```python
model1R.compile(
    loss= tf.keras.losses.MeanSquaredError(),
    optimizer=tf.keras.optimizers.Adam(0.01),
    metrics=["accuracy"],
)

Epochs = 1000
t=time.time()
history1R = model1R.fit(train_dataset, epochs = Epochs, validation_data=val_dataset, callbacks=[callback, cp_callback_1R], verbose=0)
print("Training time:" + str(time.time()-t))
```

## Load best model

```python
model1R.load_weights(checkpoint_path_1R)
```

## Plot the loss and save it as a .txt file

```python
#os.remove("PP_HH_Loss.txt")
f = open("PP_HH_Loss.txt", "a")
for i in history1R.history["loss"]:
        f.write(str(i))
        f.write("\n")
f.close()

#os.remove("PP_HH_Loss.txt")
f = open("PP_HH_Val.txt", "a")
for i in history1R.history["val_loss"]:
        f.write(str(i))
        f.write("\n")
f.close()

plt.plot(history1R.history["loss"], label="Training loss")
plt.plot(history1R.history["val_loss"], label="Validation loss")
plt.yscale("log")
plt.legend(fontsize=17)
plt.show()
```

# Fast Lyapunov Indicator

```python
def df1R(x):

    eps = 1e-5
    
    J = np.zeros([2, 2])

    for i in range(2):
        
        x1 = x.copy()
        x2 = x.copy()

        x1[0][i] += eps
        x2[0][i] -= eps

        f1 = model1R(x1).numpy()[0]
        f2 = model1R(x2).numpy()[0]

        J[ : , i] = (f1 - f2) / (2 * eps)

    return J


def df2R(x):

    eps = 1e-5
    
    J = np.zeros([2, 2])

    for i in range(2):
        
        x1 = x.copy()
        x2 = x.copy()

        x1[0][i] += eps
        x2[0][i] -= eps

        f1 = model2R(x1).numpy()[0]
        f2 = model2R(x2).numpy()[0]

        J[ : , i] = (f1 - f2) / (2 * eps)

    return J

def FLI_1R(x0, v0, t):
    
    x=x0
    v=[v0]
    
    for k in range(t):
        x = model1R(x)
        v.append(np.matmul(df1R(x.numpy()), v[-1]))
        
    vk = []
    for vv in v:
        vk.append(np.log(np.linalg.norm(vv)))
    return np.amax(vk)



def FLI_2R(x0, v0, t):
    
    x=x0
    v=[v0]
    
    for k in range(t):
        x = model2R(x)
        v.append(np.matmul(df2R(x.numpy()), v[-1]))
        
    vk = []
    for vv in v:
        vk.append(np.log(np.linalg.norm(vv)))
    return np.amax(vk)
```

# Starting points

```python
grid = True

if grid == True:
    
    limx = 75
    limy = 75
    spacing = 30
    
    u0 = []
    for i in np.linspace(-limx, limx, spacing):
        for j in np.linspace(-limy, limy, spacing):
            u0.append([i/100, j/100])
            
else:
    Npoints = 500
    r = .5
    u0 = []

    while len(u0) < Npoints:

        x = rd.random()*rd.sample([-1, 1], 1)[0]
        y = rd.random()*rd.sample([-1, 1], 1)[0]
        if x*x + y*y < r*r:
            u0.append([x, y])
```

```python
import time
start_time = time.time()
FLI_1R(np.array([[0.1, 0.1]]), np.array([0., 1.]), 10)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
FLI_u01_1R = [FLI_1R(np.array([u]), np.array([1., 0.]), 10) for u in u0]
FLI_u00_1R = [FLI_1R(np.array([u]), np.array([.0, 1.]), 10) for u in u0]
```
