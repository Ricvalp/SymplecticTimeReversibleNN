### Standard Map

```python
def Phi(x):
    return np.array([(x[0] + x[1] -(0.1/np.pi)*np.sin(2*np.pi*x[0])) % 1, (x[1] - (0.1/np.pi)*np.sin(2*np.pi*x[0])) % 1])



Npoints = 500
u0 = []

while len(u0) < Npoints:

    x = rd.random()
    y = rd.random()
    u0.append([x, y])

T = []
for u in u0:
    T.append(Phi(u))
```

```python
plt.figure(figsize=(10,10))
#plt.rc('text', usetex=True)
#plt.rc('font', family='serif')
#plt.title(r'Dataset Perturbed Pendulum', fontsize = 25)
plt.scatter([u0[i][0] for i in range(len(u0))], [u0[i][1] for i in range(len(u0))], label = "x", linewidth = 0, color ='r', s=10)
plt.scatter([T[i][0] for i in range(len(u0))], [T[i][1] for i in range(len(u0))], label = "T(x)", linewidth = 0, color = 'b', s=10)
plt.xlabel("q", fontsize=28, labelpad=8)
plt.ylabel("p", fontsize=28, labelpad=15)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid(axis='both', alpha=.3)
plt.legend(fontsize = 25)
#plt.xlim(0.1,0.8)
#plt.ylim(-0.9,0.9)
#plt.savefig("DoublePendulum_Example")
plt.show()
```
