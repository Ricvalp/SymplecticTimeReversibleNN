```python
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
file_nameHH_L = "PP_HH_Loss.txt"
file_nameSN_L = "PP_SN_Loss.txt"
file_nameNN_L = "PP_NN_Loss.txt"
file_nameRNN_L = "PP_RNN_Loss.txt"

file_nameHH_V = "PP_HH_Val.txt"
file_nameSN_V = "PP_SN_Val.txt"
file_nameNN_V = "PP_NN_Val.txt"
file_nameRNN_V = "PP_RNN_Val.txt"


file = open(file_nameHH_L, "r")
line_count = 0
for line in file:
    if line != "\n":
        line_count += 1
file.close()


Nlines = int(line_count/2)


####### HH ##########


PP_HH_loss = []

prova = open(file_nameHH_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    PP_HH_loss.append(x)
prova.close()

PP_HH_val = []

prova = open(file_nameHH_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    PP_HH_val.append(x)
prova.close()

######## SN ########

PP_SN_loss = []

prova = open(file_nameSN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    PP_SN_loss.append(x)
prova.close()

PP_SN_val = []

prova = open(file_nameSN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    PP_SN_val.append(x)
prova.close()

######### NN #############

PP_NN_loss = []

prova = open(file_nameNN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    PP_NN_loss.append(x)
prova.close()

PP_NN_val = []

prova = open(file_nameNN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    PP_NN_val.append(x)
prova.close()


######### RNN #############

PP_RNN_loss = []

prova = open(file_nameRNN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    PP_RNN_loss.append(x)
prova.close()

PP_RNN_val = []

prova = open(file_nameRNN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    PP_RNN_val.append(x)
prova.close()
```

```python
############## PLOT LOSS ##################

plt.figure(figsize = (13,6))
plt.rc('font', family='serif')
plt.title(r'Perturbed Pendulum', fontsize = 30)

plt.plot(PP_NN_loss, label="Training loss $\mathbf{H}'_{NN}$", linewidth = 3)
plt.plot(PP_NN_val, label="Validation loss $\mathbf{H}'_{NN}$", linewidth = 3)

plt.plot(PP_SN_loss, label="Training loss $\mathbf{H}'_{SN}$", linewidth = 3)
plt.plot(PP_SN_val, label="Validation loss $\mathbf{H}'_{SN}$", linewidth = 3)

plt.plot(PP_RNN_loss, label="Training loss $\mathbf{H}'_{R}$", linewidth = 3)
plt.plot(PP_RNN_val, label="Validation loss $\mathbf{H}'_{R}$", linewidth = 3)

plt.plot(PP_HH_loss, label="Training loss $\mathbf{H}'_{HR}$", linewidth = 3)
plt.plot(PP_HH_val, label="Validation loss $\mathbf{H}'_{HR}$", linewidth = 3)


plt.xlabel(r'Epoch', fontsize=20)
plt.ylabel(r'Loss', fontsize=20)
plt.yscale("log")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=15)
plt.grid(axis='both', alpha=.3)
plt.savefig("Loss")
```
