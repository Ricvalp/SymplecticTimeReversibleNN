```python
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
file_nameHH_L = "HH_Loss.txt"
file_nameSN_L = "SN_Loss.txt"
file_nameNN_L = "NN_Loss.txt"
file_nameRNN_L = "RNN_Loss.txt"

file_nameHH_V = "HH_Val.txt"
file_nameSN_V = "SN_Val.txt"
file_nameNN_V = "NN_Val.txt"
file_nameRNN_V = "RNN_Val.txt"


file = open(file_nameHH_L, "r")
line_count = 0
for line in file:
    if line != "\n":
        line_count += 1
file.close()


Nlines = int(line_count)


####### HH ##########


HH_loss = []

prova = open(file_nameHH_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    HH_loss.append(x)
prova.close()

HH_val = []

prova = open(file_nameHH_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    HH_val.append(x)
prova.close()

######## SN ########

SN_loss = []

prova = open(file_nameSN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    SN_loss.append(x)
prova.close()

SN_val = []

prova = open(file_nameSN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    SN_val.append(x)
prova.close()

######### NN #############

NN_loss = []

prova = open(file_nameNN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    NN_loss.append(x)
prova.close()

NN_val = []

prova = open(file_nameNN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    NN_val.append(x)
prova.close()


######### RNN #############

RNN_loss = []

prova = open(file_nameRNN_L, "r")
for i in range(Nlines):
    x = float(prova.readline())
    RNN_loss.append(x)
prova.close()

RNN_val = []

prova = open(file_nameRNN_V, "r")

for i in range(Nlines):
    x = float(prova.readline())
    RNN_val.append(x)
prova.close()
```

```python
############## PLOT LOSS ##################

plt.figure(figsize = (13,6))
plt.rc('font', family='serif')
plt.title(r'H$\'e$non Heiles', fontsize = 30)

plt.plot(NN_loss, label="Training loss $\mathbf{H}'_{NN}$", linewidth = 3)
plt.plot(NN_val, label="Validation loss $\mathbf{H}'_{NN}$", linewidth = 3)

plt.plot(SN_loss, label="Training loss $\mathbf{H}'_{SN}$", linewidth = 3)
plt.plot(SN_val, label="Validation loss $\mathbf{H}'_{SN}$", linewidth = 3)

plt.plot(RNN_loss, label="Training loss $\mathbf{H}'_{R}$", linewidth = 3)
plt.plot(RNN_val, label="Validation loss $\mathbf{H}'_{R}$", linewidth = 3)

plt.plot(HH_loss, label="Training loss $\mathbf{H}'_{HR}$", linewidth = 3)
plt.plot(HH_val, label="Validation loss $\mathbf{H}'_{HR}$", linewidth = 3)


plt.xlabel(r'Epoch', fontsize=20)
plt.ylabel(r'Loss', fontsize=20)
plt.yscale("log")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=15)
plt.grid(axis='both', alpha=.3)
plt.savefig("Loss_HH")
```

### Plot error

```python
file_nameHM = "HM_error.txt"
file_nameSN = "SN_error.txt"
file_nameR = "R_error.txt"


file = open(file_nameHM, "r")
line_count = 0
for line in file:
    if line != "\n":
        line_count += 1
file.close()


Nlines = int(line_count)

####### HH ##########

HM_error = []

prova = open(file_nameHM, "r")
for i in range(Nlines):
    x = float(prova.readline())
    HM_error.append(x)
prova.close()

######## SN ########

SN_error = []

prova = open(file_nameSN, "r")
for i in range(Nlines):
    x = float(prova.readline())
    SN_error.append(x)
prova.close()

######## R ########

R_error = []

prova = open(file_nameR, "r")
for i in range(Nlines):
    x = float(prova.readline())
    R_error.append(x)
prova.close()
```

```python
plt.figure(figsize=(10,10))
plt.rc('font', family='serif')
plt.title(r'H$\'e$non Heiles', fontsize = 30)
plt.plot([i for i in range(1,51)], SN_error, label="$\mathbf{H}'_{SN}$", linewidth = 4)
plt.plot([i for i in range(1,51)], R_error, label="$\mathbf{H}'_{R}$", linewidth = 4)
plt.plot([i for i in range(1,51)], HM_error, label="$\mathbf{H}'_{HR}$", linewidth = 4)

plt.xlabel(r'$n$', fontsize=28, labelpad=8)
plt.ylabel(r'Error', fontsize=28, labelpad=15)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid(axis='both', alpha=.3)
lgnd = plt.legend(loc="upper right", numpoints=1, fontsize=20)
lgnd.legendHandles[0]._sizes = [35]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
plt.savefig("Error_Test_HH_HM")
plt.show()
```
