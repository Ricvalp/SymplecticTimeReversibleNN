```julia
using NBodySimulator
using StaticArrays
using Plots
```

```julia
body1 = MassBody(SVector(0., 0., 0.), SVector(0., 0., 0.), 100.0)
body2 = MassBody(SVector(0.1, 0., 0.), SVector(0., 0.01, 0.), 10.0)
body3 = MassBody(SVector(0.0, 0.0, 0.0), SVector(-0.0001412, -0.0001412, 0.), 2.0)
```

```julia
const G = 6.67e-8 # m^3/kg/s^2
system = GravitationalSystem([body1, body2], G)
```

```julia
tspan = (0.0, 400.0)
simulation = NBodySimulation(system, tspan)
sim_result = run_simulation(simulation, Tsit5(), saveat= 1., abstol=1e-10, reltol=1e-10)
sim_result.solution
```

```julia
animate(sim_result, "path_to_animated_particles.gif")
```
